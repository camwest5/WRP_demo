## Using Markdown



Some of the basics include
________

Asterisk * for bold and italics

*one asterisk for italics*

**two asterisks for bold**

***three for both***

________

Tilde ~ for strikethrough

~like the asterisks, these need to be on either side~

________

Dollar signs $ for mathematics (LaTeX)

A single dollar sign on each side make this equation, $ F = ma $, an inline one.

Using two will give it a new line and centre it:

$$ G_{\mu\nu} + \Lambda g_{\mu\nu} = \kappa T_{\mu\nu} $$

The code inside dollar signs isn't Markdown anymore - it's $\LaTeX$.

### Data Analysis

We can use the package pandas to perform data analysis like we did in R.

To import data from a .csv file we use `pd.read_csv` (pd points Python to the pandas module)

```python
# Import data

df = pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
```

Look at the top and bottom rows with `df.head()` and `df.tail()`

```python
df.head()
```

```python
df.tail()
```

Filter for just the Australia rows:

```python
df[df.country == "Australia"]
```

Plot the population over time

```python
df[df.country == "Australia"].plot(x = "year", y = "pop")
```
